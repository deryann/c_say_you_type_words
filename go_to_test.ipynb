```python
import pyttsx3
engine = pyttsx3.init()

# 列出可用的聲音
voices = engine.getProperty('voices')
for index, voice in enumerate(voices):
    print(f"{index}: {voice.name}")

```

```python
import random

def speak_word(word):
    # 從1,2 隨便選一個
    id_random_012 = random.randint(1, 2)
    engine.setProperty('voice', voices[id_random_012].id)

    # 語音合成
    engine.say(word)
    engine.runAndWait()

speak_word(word)
```
